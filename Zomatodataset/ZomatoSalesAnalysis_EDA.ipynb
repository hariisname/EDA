## Making rest Types less than 1000 in frequency as other

```python
rest_types=df['rest_type'].value_counts(ascending=False)
```

```python
rest_types
```

```python
rest_type_lessthan1000=rest_types[rest_types<1000]
rest_type_lessthan1000
```

## Making rest_type less than 1000 in frequency as Others

```python
def handle_rest_type(value):
    if(value in rest_type_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply( handle_rest_type)
df['rest_type']
```

```python
df['rest_type'].value_counts()
```

```python
df.head()
```

## Cleaning Location column

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)

location_lessthan300=location[location<300]

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df['location']=df['location'].apply( handle_location)
df['location']
```

## Cleaning cuisines Column

```python
df['cuisines'].value_counts()
```

```python
cuisines=df['cuisines'].value_counts(ascending=False)

cuisines_lessthan100=cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply( handle_cuisines)
df['cuisines']
```

#### Univariate Plots:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for 'rate'
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='rate', bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Bar plot for 'online_order'
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='online_order')
plt.title('Count of Restaurants Accepting Online Orders')
plt.xlabel('Online Order Availability')
plt.ylabel('Count')
plt.show()

# Pie chart for 'rest_type'
plt.figure(figsize=(8, 8))
df['rest_type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Restaurant Types')
plt.ylabel('')
plt.show()

# Box plot for 'votes'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='votes')
plt.title('Distribution of Votes')
plt.ylabel('Votes')
plt.show()
```
